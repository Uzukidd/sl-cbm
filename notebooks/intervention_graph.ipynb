```python
import torch
import pandas as pd

def to_txt(x, data:torch.Tensor, method:str):
    mean = data.mean(dim=0)
    std = data.std(dim=0)

    df = pd.DataFrame({
        'x': x,
        'mean': mean.numpy(),
        'std': std.numpy()
    })

    df.to_csv(f".temp/{method}_intervention.txt", sep='\t', index=False, float_format='%.4f')
```

## Intervention over RIVAL10

```python
intervetion_data = torch.load("../outputs/evals/PCBM_RN18_CUB/evaluations/intervention_taskerror.pt")

for name, data in intervetion_data.items():
    to_txt(range(0, 111, 10), data, name)
```
